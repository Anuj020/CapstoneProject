```python
# Define a list of menu items
menu_items = ['Big Mac', 'Quarter Pounder with Cheese', 'Double Quarter Pounder with Cheese', 'McChicken', 'Filet-O-Fish', 'Chicken McNuggets', 'McDouble', 'Cheeseburger', 'Hamburger', 'French Fries', 'Poutine', 'Onion Rings', 'Coca-Cola', 'Sprite', 'Fanta', 'Iced Coffee', 'Hot Chocolate', 'Latte', 'Cappuccino']
prices = ['4.99', '5.99', '6.99', '4.49', '4.29', '6.99', '2.39', '1.69', '1.39', '1.79', '3.99', '3.49', '1.89', '1.89', '1.89', '2.19', '2.39', '2.99', '3.49', '3.99']

# Define the range for the random values
low, high = 1, 10

# Add a new column "menu_item" with random menu item values to the dataframe
df_mcdonalds['menu_item'] = np.random.choice(menu_items, size=len(df_mcdonalds))

# View the updated dataset
df_mcdonalds.head()
```

```python
# Define a dictionary that maps menu items to prices
menu_item_prices = {'Big Mac': 4.99, 
                    'Quarter Pounder with Cheese': 5.99, 
                    'Double Quarter Pounder with Cheese': 6.99, 
                    'McChicken': 4.49, 
                    'Filet-O-Fish': 4.29, 
                    'Chicken McNuggets': 6.99, 
                    'McDouble': 2.39, 
                    'Cheeseburger': 1.69, 
                    'Hamburger': 1.39, 
                    'French Fries': 1.79, 
                    'Poutine': 3.99, 
                    'Onion Rings': 3.49, 
                    'Coca-Cola': 1.89, 
                    'Sprite': 1.89, 
                    'Fanta': 1.89, 
                    'Iced Coffee': 2.19, 
                    'Hot Chocolate': 2.39, 
                    'Latte': 2.99, 
                    'Cappuccino': 3.49}

# Add a new column "price" that maps to the menu_item_prices dictionary
df_mcdonalds['price'] = df_mcdonalds['menu_item'].map(menu_item_prices)

# View the updated dataset
df_mcdonalds.head()
```

```python
df_mcdonalds.drop(['latitude','longitude','postalCode','id','dateAdded','dateUpdated','keys'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


cat= ['address','categories','city','country','name','province']


# Encode the categorical variables using one-hot encoding
df_encoded = pd.get_dummies(df_mcdonalds, columns=cat)

df_encoded 
```

```python
from sklearn.model_selection import train_test_split

feature = df_encoded.drop('menu_item',axis=1)
target = df_encoded['menu_item']

X_train, X_test, y_train, y_test = train_test_split(feature,target,test_size=0.3)

print("X_train:", X_train.shape)

print("y_train:",y_train.shape)

print("--------------------------------------------")

print("X_test:",X_test.shape)

print("y_test:",y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# train a decision tree classifier
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train,y_train)

# make prediction on the testing data
y_prediction = model.predict(X_test)
print("Prediction: ",y_prediction)

# evaluate the performance of the classifier
accuracy = accuracy_score(y_test,y_prediction)
print("Accuracy:",accuracy*100)
```

```python
from sklearn.model_selection import cross_val_score, 

score = cross_val_score(model, feature, target,cv = 50)

print("Cross validation score: ",score.mean())
```
